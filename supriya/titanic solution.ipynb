```python
data = pd.read_csv('titanic.csv')
```

```python
data.describe()
```

```python
sns.lmplot(x= 'Survived', y= 'Age', data = data)
```

```python
sns.boxplot(data = data)
```

```python
sns.violinplot(x = 'Pclass', y = 'Age', data = data)
plt.xticks(rotation = -45)
```

```python
sns.swarmplot(x = 'Pclass', y = 'Age', data = data)
plt.xticks(rotation = -45)
```

```python
#Pandas slicing the dataframe.
col = list(data.columns)
stats_df = data[col[4:-2]]
#Correlation of the different varibles can be known by ".corr()" function on pandas dataframe.
corr = stats_df.corr()
sns.heatmap(corr)
```

```python
sns.distplot(data['Survived'], kde= False)
```

```python
sns.countplot(x = 'Pclass', data = data)
plt.xticks(rotation = -45)
```
