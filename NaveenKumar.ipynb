```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
tit = pd.read_csv("titanic.csv",index_col=0)
```

```python
tit.head()
```

```python
sns.lmplot(row= 'Sex', col= 'Survived', data = tit,x='Age',y='Parch')
```

```python
tit.describe()
```

```python
sns.barplot(data = tit,x="Pclass",y="Survived")
```

```python
tit.info()
```

```python
sns.barplot(x="Sex",y="Survived",data=tit)
```

```python
sns.swarmplot(x = 'Sex', y = 'Fare', data = tit)
plt.xticks(rotation = -45)
```

```python
col=list(tit.columns)
d=tit[col[1:-1]]
corr=d.corr()
sns.heatmap(corr)
```

```python
sns.distplot(tit['Fare'],kde=False)
```

```python
sns.boxplot(data=tit,x='Pclass',y='Fare')
```

```python
sns.violinplot(x = 'Embarked', y = 'Fare', data = tit)
plt.xticks(rotation = -45)
```

```python
sns.countplot(x = 'Embarked', data = tit)
plt.xticks(rotation = -39)
```

```python
g = sns.FacetGrid(tit, col='Survived', row='Pclass')
g.map(plt.hist, 'Age')
g.add_legend();
```
