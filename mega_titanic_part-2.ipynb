# 4>Feature Engineering

is basically "Fill the missing values in the obtained data"

```python
plt.figure(figsize=(12, 7))
sns.boxplot(x='Pclass',y='Age',data=data,palette='winter')
```

```python
data.groupby('Pclass')['Age'].median()
```

#by considering the median of the age of different categories of classes we fill the white lines in the matrix or empty cells  or NaN values

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        # Class-1
        if Pclass == 1:
            return 37
        # Class-2 
        elif Pclass == 2:
            return 29
        # Class-3
        else:
            return 24

    else:
        return Age

```

```python
data['Age'] = data[['Age','Pclass']].apply(impute_age,axis=1)
```

```python
ms.matrix(data)
```

```python
#Let's drop the Cabin column and the row in the Embarked that is NaN.
```

```python
data.drop('Cabin', axis = 1,inplace=True)
```

```python
data.dropna(inplace = True)
```

```python
ms.matrix(data)
```

```python
data.info()
```

```python
data['Sex'].unique()
```

```python
data['Sex'].value_counts()
```

```python
sex_df = pd.get_dummies(data['Sex'],drop_first=3)
sex_df.head()
```

```python
data['Embarked'].unique()
```

```python
data['Embarked'].value_counts()
```

```python
embark_df = pd.get_dummies(data['Embarked'],drop_first=True)
embark_df.head()
```

```python
old_data = data.copy()
data.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
data.head()
```

```python
old_data.info()
```

```python
data = pd.concat([data,sex_df,embark_df],axis=1)
```

```python
data.info()
```

# 5>Model Selection

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.drop('Survived',axis=1), 
                                                    data['Survived'], test_size=0.30, 
                                                    random_state=101)
```

```python
#trian the data(logical regression)
from sklearn.linear_model import LogisticRegression

# Build the Model.
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predicted =  logmodel.predict(X_test)
```

```python
predicted
```

# 6>Tune The Model

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(confusion_matrix(y_test, predicted))
```

```python
#precision score
from sklearn.metrics import precision_score

print(precision_score(y_test,predicted))
```

```python
#recall score
from sklearn.metrics import recall_score

print(recall_score(y_test,predicted))
```

```python
#f1_score
from sklearn.metrics import f1_score

print(f1_score(y_test,predicted))
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test,predicted))
```
