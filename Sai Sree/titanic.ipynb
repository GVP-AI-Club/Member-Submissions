```python
#fit_reg=True, hue='Stage'
sns.lmplot(x= 'Pclass', y= 'Survived', data = ti)
```

```python
sns.boxplot(data = ti)
```

```python
sns.violinplot(x = 'Age', y = 'Survived', data = ti)
plt.xticks(rotation = -45)
```

```python
sns.swarmplot(x = 'Age', y = 'Survived', data = ti)
plt.xticks(rotation = -45)
```

```python
#Pandas slicing the dataframe.
col = list(ti.columns)
stats_df = ti[col[4:-2]]
#Correlation of the different varibles can be known by ".corr()" function on pandas dataframe.
corr = stats_df.corr()
sns.heatmap(corr)
```

```python
sns.distplot(ti['Survived'], kde= False)
```

```python
sns.countplot(x = 'Age', data = ti)
plt.xticks(rotation = -45)
```
