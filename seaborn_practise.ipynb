```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
ti = pd.read_csv('titanic.csv')
```

```python
ti.head()
```

```python
type(ti)
```

```python
sns.lmplot(x= 'Age', y= 'Pclass', data= ti.head()) #x and y must be assingned to cols containing values
```

```python
sns.boxplot(data = ti.head()) #only numbered columns are plotted
```

```python
sns.boxplot(data = ti.tail())   #as passenger id is more compared to other
```

```python
sns.violinplot(data= ti.head())
```

```python
sns.violinplot(data= ti)
```

```python
sns.violinplot(x= 'Sex', y='Age', data= ti.head()) 
```

```python
sns.swarmplot(x= 'Age', y='Name', data=ti.head())
```

```python
sns.catplot(x='Pclass', y='Age', kind='swarm', data=ti)
```

```python
sns.swarmplot(y= 'Age', x='Pclass', hue="Sex", data=ti)
```
