```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pok = pd.read_csv('titanic.csv', index_col=0, sep = ',')
```

```python
pok.head()
```

```python
#fit_reg=True, hue='stage'
sns.lmplot(x= 'Age', y= 'Pclass', data = pok)
```

```python
sns.lmplot(x= 'Age', y= 'Pclass', data = pok)
```

```python
sns.boxplot(data = pok)
```

```python
sns.violinplot(x = 'SibSp', y = 'Survived', data = pok)
plt.xticks(rotation = -45)
```

```python
sns.swarmplot(x = 'Fare', y = 'Cabin', data = pok)
plt.xticks(rotation = -45)
```

```python
sns.swarmplot(x = 'Survived', y = 'Fare', data = pok)
plt.xticks(rotation = -45)
```

```python
#Pandas slicing the dataframe.
col = list(pok.columns)
stats_df = pok[col[4:-2]]
#Correlation of the different varibles can be known by ".corr()" function on pandas dataframe.
corr = stats_df.corr()
sns.heatmap(corr)
```

```python
sns.distplot(pok['Survived'], kde= False)
```

```python
sns.countplot(x = 'Age', data = pok)
plt.xticks(rotation = -45)
```

```python
sns.countplot(x = 'Survived', data = pok)
plt.xticks(rotation = -45)
```
