```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
titan=pd.read_csv('titanic.csv', index_col=0, sep = ',')
```

```python
sns.violinplot(x='Pclass',y='Age',data=titan)
```

```python
sns.catplot(x = 'Sex', y = 'Pclass',kind="violin", hue="Survived",split=True,data = titan)
```

```python
sns.violinplot(x = 'Sex', y = 'Pclass', hue="Survived",data = titan)
```

```python
g = sns.catplot(x="Survived", y="Age", kind="violin", inner=None, data=titan)
sns.swarmplot(x="Survived", y="Age", color="k", data=titan);
```

```python
sns.catplot(x="Sex", y="Survived", hue="Pclass", kind="point", data=titan);
```

```python
sns.countplot(x = 'Pclass', data = titan)
```

```python
sns.catplot(x="Sex", y="Survived", hue="Pclass", kind="bar", data=titan);
```

```python
sns.countplot(x = 'Age', data = titan)
```

```python
sns.catplot(x="Embarked", y="Pclass", hue="Survived",
            col="Sex",aspect=.5,
            kind="swarm", data=titan);
```

```python
sns.catplot(x="Fare", y="Survived", row="Pclass",
                kind="box", orient="h" ,height=1.5, aspect=4,
                data=titan)
```

```python
sns.lmplot(x='Fare',y='Age',data=titan)
```
