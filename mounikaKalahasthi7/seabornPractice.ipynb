```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
myFile = pd.read_csv('titanic.csv',sep=',', index_col= 0)
```

```python
myFile.head()
```

```python
#Scatter plot
```

```python
sb.lmplot(x='Age',y='Fare',data=myFile, fit_reg= True, hue = 'Sex')
```

```python
#Box plot
```

```python
sb.boxplot(data= myFile)
```

```python
sb.boxplot(x='Sex', y='Age', data=myFile)
```

```python
#Violin plot
```

```python
sb.violinplot(data=myFile)
```

```python
sb.violinplot(x='Fare', data=myFile)
```

```python
sb.violinplot(y='Fare', data=myFile)
```

```python
sb.violinplot(x='Sex',y='Fare', data=myFile)
plt.xticks(rotation = +45)
```

```python
sb.swarmplot(x='Sex',y='Fare', data=myFile)
```

```python
sb.swarmplot(y='Fare', data=myFile)
```

```python
#heatmap
```

```python
cols = list(myFile.columns)
select = cols[4:6]
select.append(cols[8])
selectData = myFile[select]
corr = selectData.corr()
sb.heatmap(corr)
```
