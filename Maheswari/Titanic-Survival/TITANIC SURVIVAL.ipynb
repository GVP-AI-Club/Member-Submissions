```python
from scipy import stats
data.groupby(['Pclass','Sex'])['Age'].apply(stats.mode)
```

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    Sex = cols[2]
    
    
    if pd.isnull(Age):
        # Class-1
        
      
        if Pclass == 1:
            if Sex == 'male':
              return 36
            if Sex == 'female':
              return 35
        # Class-2 
        elif Pclass == 2:
            if Sex == 'male':
              return 23
            if Sex == 'female': 
              return 24
        # Class-3
        else:
            if Sex == 'male':
              return 22
            if Sex == 'female':
              return 18

    else:
        return Age
```

```python
data['Age'] = data[['Age','Pclass','Sex']].apply(impute_age,axis=1)
```

```python
data.drop('Cabin', axis = 1,inplace=True)
```

```python
data.dropna(inplace = True)
```

```python
data.head()
```

```python
sex_df = pd.get_dummies(data['Sex'],drop_first=3)
sex_df.head(3)
```

```python
embark_df = pd.get_dummies(data['Embarked'],drop_first=True)
embark_df.head()
```

```python
old_data = data.copy()
data.drop(['Sex','Embarked','Name','Ticket'],axis=1,inplace=True)
data.head()
```

```python
data = pd.concat([data,sex_df,embark_df],axis=1)
```

```python
data.head(2)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data.drop('Survived',axis=1), 
                                                    data['Survived'], test_size=0.30, 
                                                    random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression

# Build the Modelfr.
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

```python
predicted=logmodel.predict(X_test)
```
